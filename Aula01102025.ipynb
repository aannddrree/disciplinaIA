```python
import csv

with open('dados.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['Nome', 'Idade', 'Cidade'])
    writer.writerow(['Jo達o', '25', 'S達o Paulo'])
    writer.writerow(['Maria', '30', 'Rio de Janeiro'])
    writer.writerow(['Pedro', '28', 'Belo Horizonte'])

print('Arquivo "dados.csv" criado com sucesso!')
```

```python
with open ("dados.csv", 'r') as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)
```
