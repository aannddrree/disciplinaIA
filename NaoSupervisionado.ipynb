Aprendizado Não Supervisionado

No aprendizado não supervisionado, a máquina recebe os dados sem resposta pronta e tenta encontrar os padrões sozinha.

Exemplo:

- Separar clientes parecidos
- Descobrir grupos de música parecidos
- Identificar perfis semelhantes de consumo (cliente)

```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Dados de altura
dados = np.array([
    [150],
    [155],
    [160],
    [170],
    [175],
    [180]
])

# criação de um modelo para 2 grupos
modelo_kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)

# treinamento
modelo_kmeans.fit(dados)

# Mostrar os grupos encontrados
print("Grupos encontrados")
print(modelo_kmeans.labels_)

print("Centroides dos grupos")
print(modelo_kmeans.cluster_centers_)
```

```python
#Visualizar os grupos

plt.figure(figsize=(8,5))
plt.scatter(dados[:,0], [1]*len(dados), c=modelo_kmeans.labels_, s=120)
plt.xlabel("Altura")
plt.yticks([])
plt.title("Altura dos clientes - Aprendizado não supervisionado")
plt.grid(True)
plt.show()
#
```
