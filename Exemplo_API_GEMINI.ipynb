# **Import das bibliotecas**

```python
#Importanto das bibliotecas necessarias para poder chamar o servico do Gemini
import google.generativeai as genai
```

# Exemplo para chamar a API do Gemini utilizando python *

https://aistudio.google.com/app/api-keys

```python
#chave de API
genai.configure(api_key="xxx")

# Modelo padrão
MODEL_NAME = "gemini-2.5-flash"

# Listar modelos disponíveis
for m in genai.list_models():
    print(m.name)
```

```python
# Funcao para chamar o servico do GEMINI:
def generate_text(prompt: str):
    try:
        model = genai.GenerativeModel(MODEL_NAME)
        response = model.generate_content(prompt)
        return {"response": response.text}
    except Exception as e:
        return {"error": str(e)}

# Enviar pergunta para IA
result = generate_text("Vai chover?")
print(result)
```
