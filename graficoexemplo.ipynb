```python
import matplotlib.pyplot as plt
import numpy as np

# Gerar dados de exemplo
dias = np.linspace(0, 1000, 500)

# Funções de sobrevida (exemplo)
sobrevida1 = np.exp(-0.001 * dias)  # Sobrevida exponencial decrescente
sobrevida2 = np.exp(-0.0005 * dias)  # Sobrevida exponencial com decaimento mais lento
sobrevida3 = np.exp(-0.0008 * dias)  # Sobrevida exponencial com decaimento intermediário

# Criar o gráfico
plt.figure(figsize=(10, 6))
plt.plot(dias, sobrevida1, label='Grupo 1', color='blue')
plt.plot(dias, sobrevida2, label='Grupo 2', color='green')
plt.plot(dias, sobrevida3, label='Grupo 3', color='red')

# Adicionar título e rótulos
plt.title('Gráfico de Sobrevida')
plt.xlabel('Tempo (dias)')
plt.ylabel('Sobrevida')
plt.legend()

# Exibir o gráfico
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Gerar dados de exemplo
dias = np.linspace(0, 1000, 500)

# Funções de sobrevida (exemplo)
sobrevida1 = np.exp(-0.001 * dias)  # Sobrevida exponencial decrescente
sobrevida2 = np.exp(-0.0005 * dias)  # Sobrevida exponencial com decaimento mais lento
sobrevida3 = np.exp(-0.0008 * dias)  # Sobrevida exponencial com decaimento intermediário

# Simular censuras
np.random.seed(42)
censuras1 = np.random.choice(dias, size=15, replace=False)
censuras2 = np.random.choice(dias, size=15, replace=False)
censuras3 = np.random.choice(dias, size=15, replace=False)

# Criar o gráfico
plt.figure(figsize=(10, 6))
plt.plot(dias, sobrevida1, label='Grupo 1', color='blue')
plt.plot(dias, sobrevida2, label='Grupo 2', color='green')
plt.plot(dias, sobrevida3, label='Grupo 3', color='red')

# Adicionar censuras
plt.scatter(censuras1, np.exp(-0.001 * censuras1), color='blue', edgecolor='black', zorder=5)
plt.scatter(censuras2, np.exp(-0.0005 * censuras2), color='green', edgecolor='black', zorder=5)
plt.scatter(censuras3, np.exp(-0.0008 * censuras3), color='red', edgecolor='black', zorder=5)

# Adicionar título e rótulos
plt.title('Gráfico de Sobrevida com Censuras')
plt.xlabel('Tempo (dias)')
plt.ylabel('Sobrevida')
plt.legend()

# Exibir o gráfico
plt.grid(True)
plt.show()
```
